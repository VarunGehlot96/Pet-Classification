```python
# importing libraries

import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img, img_to_array
```

```python
# Checking TensorFlow Version 

tf.__version__
```

```python
# Data Preprocessing

train_datagen = ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True)
training_set = train_datagen.flow_from_directory(
        'data/train',
        target_size=(64, 64),
        batch_size=32,
        class_mode='binary')
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)
test_set = test_datagen.flow_from_directory(
        'data/test',
        target_size=(64, 64),
        batch_size=32,
        class_mode='binary')
```

```python
# Instantiating Model

cnn = tf.keras.models.Sequential()
```

```python
# Adding Convolutional Neural Network and Input

cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size= 5,activation='relu', input_shape = [64,64,3]))
```

```python
# Pooling

cnn.add(tf.keras.layers.MaxPool2D( pool_size=2 , strides=2 ))
```

```python
# Adding Second CNN layer with extra filters

cnn.add(tf.keras.layers.Conv2D(filters=64, kernel_size= 5,activation='relu'))
```

```python
cnn.add(tf.keras.layers.MaxPool2D( pool_size=2 , strides=2 ))
```

```python
# Flattening

cnn.add(tf.keras.layers.Flatten())
```

```python
cnn.add(tf.keras.layers.Dense(units= 32, activation='relu'))
```

```python
# Output layer

cnn.add(tf.keras.layers.Dense(units = 1,activation = 'sigmoid'))
```

```python
#Compiling the CNN

cnn.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# Training the CNN

cnn.fit(x = training_set , validation_data= test_set, epochs = 300)
```
